```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://thuvienphapluat.vn/van-ban/Giao-thong-Van-tai/Luat-trat-tu-an-toan-giao-thong-duong-bo-444251.aspx"
response = requests.get(url)
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    content_part = soup.find('div', class_='content1')
    if content_part:
        contents = content_part.css.select('p span')
        content = ''
        for text in contents:
            content += ' ' + text.text.strip()
        # Open the file in write mode and write the content to it
        with open("Luat-trat-tu-an-toan-giao-thong-duong-bo-444251.txt", "w", encoding="utf-8") as f:
            f.write(content)

        print("Content saved to 'Luat-trat-tu-an-toan-giao-thong-duong-bo-444251.txt'")
    else:
        print("Content not found on the page.")
else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```

```python
import os
import pandas as pd
def getContentArticle(hrefs,nameFile):
    data = {'title': [], 'embedding_title': [], 'content': [], 'source': []}
    # Check if the CSV file already exists
    if os.path.exists(nameFile + ".csv"):
        # Load the existing DataFrame
        existing_df = pd.read_csv(nameFile + ".csv")
    else:
        # Create an empty DataFrame if the file doesn't exist
        existing_df = pd.DataFrame(columns=['title', 'embedding_title', 'content', 'source'])

    for i in hrefs:
        try:
            response_article = requests.get(i)
            soup_article = BeautifulSoup(response_article.content, 'html.parser')
            content_part = soup_article.css.select('article .col-md-9')[0]
            
            if content_part:
                title = content_part.find('h1', class_='title')
                contents = content_part.css.select('section.news-content p')
                
                title_text = title.text.strip()
                title_embedding = get_embedding().embed_query(title_text)
                
                content = title.text.strip()
                for text in contents:
                    content += '\n' + text.text.strip()
                
                data['title'].append(title_text)
                data['embedding_title'].append(title_embedding)
                data['content'].append(content)
                data['source'].append(i)
                
                print("Done ", title_text[:20])
            else:
                print("Content not found on the page.")
        except Exception as e:
            print(str(e))
            continue
        
    # Combine the existing DataFrame with the new data
    new_df = pd.DataFrame(data)
    combined_df = pd.concat([existing_df, new_df], ignore_index=True)

    # Save the combined DataFrame to a CSV file
    combined_df.to_csv(nameFile + ".csv", index=False)
    return combined_df.tail(5)
```

```python
from langchain_community.embeddings import HuggingFaceBgeEmbeddings


def get_embedding():
    model_name = "BAAI/bge-m3"
    encode_kwargs = {'normalize_embeddings': True}

    bge_embeddings = HuggingFaceBgeEmbeddings(
        model_name=model_name,
        model_kwargs = {'device': 'cpu'},
        encode_kwargs=encode_kwargs
    )
    
    return bge_embeddings
```

```python
embed_text = get_embedding().embed_query("Tổ chức nước ngoài đầu tư xây dựng kết cấu hạ tầng đường thủy nội địa có cần phải tuân thủ các quy định về đầu tư công hay không?")
```

```python
print(embed_text)
```

```python
embed_query = get_embedding().embed_query("Tổ chức ở nước ngoài đầu tư xây dựng  hạ tầng đường thủy nội địa có cần tuân thủ các quy định về đầu tư công hay không?")
```

```python
from scipy.spatial.distance import cosine, euclidean
cosine_score = 1 - cosine(embed_text, embed_query)
print("cosine", cosine_score)
euclidean_score = euclidean(embed_text, embed_query)
print("euclidean", euclidean_score)
```

```python
from urllib.parse import urlparse
import requests
from bs4 import BeautifulSoup
# Pháp luật về Giao thông - Vận tải
url3 = "https://thuvienphapluat.vn/phap-luat/giao-thong-van-tai"
# Parse the URL
parsed_url3 = urlparse(url3)

# Extract the domain URL
domain_url3 = parsed_url3.scheme + "://" + parsed_url3.netloc
```

```python
response3 = requests.get(url3 + '?page=1')
response3
```

```python
# Pháp luật về Giao thông - Vận tải
for i in range(388, 389):
    response3 = requests.get(url3 + '?page=' + str(i))
    print('page',  i)
    print(response3.status_code)
    if response3.status_code == 200:
        soup = BeautifulSoup(response3.content, 'html.parser')
        content_part = soup.css.select('section')
        articles= content_part[0].css.select('article')
        hrefs = []
        for article in articles:
            a_tag = article.find('a', href=True)
            if a_tag:
                hrefs.append(domain_url3 + a_tag['href'])
        nameFile = "../data/gtvt2/tu-van-pha-luat-du-thao-luat-trat-tu-atgtdb"
        getContentArticle(hrefs, nameFile)
    else:
        print("Failed to retrieve the webpage. Status code:", response3.status_code)
```
